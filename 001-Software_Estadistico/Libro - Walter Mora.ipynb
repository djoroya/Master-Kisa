# Ejemplo 1.1
Es conocido que $$S=\sum_{i=1}^{\infty} \frac{1}{i^2}=\frac{\pi}{6}$$
Comprobar esto numéricamente

```r
df.error = data.frame(matrix(ncol=2))
names(df.error) <- c("k","error k")
# Valor analítico
s=pi^2/6

# Suma de los primero elementos
for ( k in seq(100,1000,50)){
x  <- 1:k
sk <- sum(1/x^2)
df.error <- rbind(df.error,c(k,s-sk))    
}    
plot(df.error,type="b")
```

# Ejercicio 1.2
Si tenemos un conjunto de datos numéricos $x = c(x_1,x_2,...,x_n)$ (una muestra) entonces su media (o promedio) es sum(x)/lenght(x), en R esta se calcula con mean(x). Supongamos que tenemos un vector x de datos (una muestra) ordenados de menor a mayor, $x = (x_1,x_2,...,x_n)$. La k−ésima muestra podada es

$$\hat{x}_k =\frac{x_{k+1} + x_{k+2} + ... + x_{n−k}}{n − 2k}$$

```r
x=c(1,3,4,6,3,1,3,4,13,4,5,4,3,4,3,4,3,2,5,7,7,5,40)
k=2

mean_trunc <- function(x,k){
    n=length(x)
    mean(sort(x)[c((k+1):(n-k))])
    }

mt1=mean_trunc(x,k)
m1=mean(x)
x=c(x,1000)
mt2=mean_trunc(x,k)
m2=mean(x)

plot(c(mt2-mt1,m2-m1),type="h",lwd=30,col="red",xlim=c(0.5,2.5),ylim=c(0,45))
```

# Ejercicio 1.3
Declarar la funcion exponencial por su serie de potencias 


```r
expo <- function(x,aprox){
    return(1+sum(cumprod(x/1:aprox)))
}

grid <- as.matrix(expand.grid(seq(-5,5,1),seq(1,10,1)))
len <- length(grid[,1])
data <- matrix(ncol=3)
for (nfila in 1:len){
    fila <- grid[nfila,]
    e <- expo(fila[[1]],fila[[2]])
    data <- rbind(data,c(grid[nfila,],e))
}
library("graphics")
library("scatterplot3d")
plot <- scatterplot3d(data, main="3D Scatterplot",highlight.3d = TRUE,type="p",xlab="x",ylab="aprox")
```
