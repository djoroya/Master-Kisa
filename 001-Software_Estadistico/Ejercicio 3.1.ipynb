Se puede trazar una circunferencia mediante puntos, es decir, por pares de números (x,y).

Estas coordenadas habituales pueden ser expresadas por coordenadas polares (radio, ángulo) mediante las siguientes relaciones:
x = r·cos(α)
y = r·sin(α) 
siendo α el ángulo, entre 0 y 2π, y r el radio.
 
Así, se puede trazar una circunferencia de radio la unidad, generando una secuencia de valores de ángulos entre 0 y 2π, que contenga, por ejemplo, 100 valores:
alpha <- seq(0,2*pi, length.out=100) 
plot(x=cos(alpha), y=sin(alpha))

 

O simplemente:
plot(cos(alpha), sin(alpha))

```r
library(ggplot2)

circu <- function(r,angl){
    x <- round(r*cos(angl),3)
    y <- round(r*sin(angl),3)
    return(matrix(c(x,y),ncol=2))
}
points <- data.frame(circu(1,seq(0,2.1*pi,0.05)))

plot(points,type="l",xlim=c(-2,2),ylim=c(-2,2))
par(new=TRUE)
points <- data.frame(circu(2,seq(0,2.1*pi,0.05)))
plot(points,type="l",xlim=c(-2,2),ylim=c(-2,2))

#plot(circu(1,seq(0,2*pi,0.3))[1:0,])
```

```r
# Generate some sample data, then compute mean and standard deviation
# in each group
df <- data.frame(
  gp = factor(rep(letters[1:3], each = 10)),
  y = rnorm(30)
)
ds <- plyr::ddply(df, "gp", plyr::summarise, mean = mean(y), sd = sd(y))

# The summary data frame ds is used to plot larger red points on top
# of the raw data. Note that we don't need to supply `data` or `mapping`
# in each layer because the defaults from ggplot() are used.
ggplot(df, aes(gp, y)) +
  geom_point() +
  geom_point(data = ds, aes(y = mean), colour = 'red', size = 3)

# Same plot as above, declaring only the data frame in ggplot().
# Note how the x and y aesthetics must now be declared in
# each geom_point() layer.
ggplot(df) +
  geom_point(aes(gp, y)) +
  geom_point(data = ds, aes(gp, mean), colour = 'red', size = 3)

# Alternatively we can fully specify the plot in each layer. This
# is not useful here, but can be more clear when working with complex
# mult-dataset graphics
ggplot() +
  geom_point(data = df, aes(gp, y)) +
  geom_point(data = ds, aes(gp, mean), colour = 'red', size = 3) +
  geom_errorbar(
    data = ds,
    aes(gp, mean, ymin = mean - sd, ymax = mean + sd),
    colour = 'red',
    width = 0.4
  )
```

```r
pdf("diana.pdf",width=6.5, height=7)
# Creamos la funcion circle que nos devuelve una matrix de dos 
# columnas donde se encuentra los circulos parametrizados
circle <-
function(radio){
    alpha <- seq(0,2*pi, length.out=50)
    curve <- matrix(c(sin(alpha),cos(alpha)),ncol=2)
    return(radio*curve)
}
# Creamos los circulos dentro de un ciclo for 
for (i in 10:2){
    if(i == 10){par(lwd=3)}else{par(lwd=1)}
    plot(circle(i),xlim=c(-10,10),ylim=c(-10,10),type="l",axes=FALSE,xlab="",ylab="")
    text(c(0),c(i)-0.3,11-i,font=2,cex=1.8)
    par(new=TRUE)
} 
# Pintamos primero el circulo solido y luego el texto "10"
i=1
polygon(circle(1), col = "red")
text(c(0),c(i)-0.3,11-i,font=2,cex=1.8)
dev.off()
```
