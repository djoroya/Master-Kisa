```r
###################################
# 
# Los datos no son homogéneos: porcentajes 
# ¿significa lo mismo (0.92-0.88) que (0.05-0.01)?
# (0.92-0.88)/0.88 <<< (0.05-0.01)/0.01
# Por tanto: estandarización
#
datos <- scale(datos0, center=TRUE, scale=TRUE)
names(datos) <- names(datos0)
#
###################################
#
# Selección del valor de 'k'. Criterios:
#                          1. Índice de bondad, 'IB'
#                          2. Índice Davies-Bouldin
#                          3. Valor medio de índice de siluetas
#                          4. Multiples índices: 'NbClust'
#
K <- 10 # número máximo de clases
#
##########
#
# 1. Índice de bondad, 'IB'
#
IB <- 0
for(k in 2:K){
  km <- kmeans(datos, k, nstart=100, iter.max=1000)
  print(km$size)
  IB[k] <- km$betweenss/km$totss
}
#
names(IB) <- 1:10
round(IB[2:K], digits=4)
# 2 3 4 5 6 7 8 9 10
# 0.3546 0.5500 0.6290 0.6908 0.7371 0.7706 0.8083 0.8315 0.8498
#
plot(2:K, IB[2:10], xlab="k", type="b", ylab="IB", las=1)
#
# Se observa que los valores k=2, k=3 e incluso k=4 parecen apropiados,
#
##########
#
```

```r
# 2. Ìndice de Davies-Bouldin
#
distancias <- dist(datos, method="euclidean", diag=TRUE, upper=TRUE)^2
distancias.m <- as.matrix(distancias, ncol=nrow(datos)) # formato necesario
#
# Definition del indice 'DB' par une fonction
#
DB <- function(distancias.m, compocl, k){
 Mcl <- matrix(rep(0,k*k), ncol=k) # initialisation
 for(kcli in 1:k){
   for(kclj in 1:k){
      Mcl[kcli,kclj] <- mean(distancias.m[compocl[[kcli]], compocl[[kclj]]])
   }
 }
 R <- matrix(rep(0,k*k), ncol=k)
 for(kcli in 1:k){
   for(kclj in 1:k){
      R[kcli,kclj] <- (Mcl[kcli,kcli]+Mcl[kclj,kclj])/Mcl[kcli,kclj]
   }
 }
 D <- rep(0,k)
 for(kcl in 1:k){
    D[kcl] <- max(R[kcl,-kcl])
 }
 DB <- mean(D)
 return(DB)
}
#
#####
#
db <- NA # db[1] no tiene sentido
for(k in 2:K){
  km <- kmeans(datos, k, nstart=100, iter.max=10000)
  compocl <- vector("list", k)
  for(kcl in 1:k) compocl[[kcl]] <- which(km$cluster == kcl)
  ( DB(distancias.m, compocl,k) )  
  db[k] <- DB(distancias.m, compocl,k)
}
#
cbind(k=1:length(db), DB=round(db, digits=6))
```

```r
#
# Representación gráfica de la variación de 'db' en función de 'k'
#
plot(2:K, db[-1], type="b", las=1, xlab="k", ylab="DB")
axis(1, at=2:10)
abline(v=2:K, col="gray")
#
# Se observa que k=10 es el mínimo.
#                k= 4 es un mínimo local.
#
##########
#
# 3. Valor medio de índice de siluetas
#
library(cluster)
#
avg.width <- NA # avg.width[1] no tiene sentido
for(k in 2:K){
  km <- kmeans(datos, k, nstart=100, iter.max=1000)
  silk <- silhouette(km$cluster, distancias)
  avg.width[k] <- summary(silk)$avg.width
}
#
cbind(k=1:length(avg.width), avg.width=round(avg.width, digits=6))
#
# Representación gráfica de la variación de 'avg.width' en función de 'k'
#
plot(2:K, avg.width[-1], type="b", las=1, xlab="k", ylab="avg.width")
axis(1, at=2:K)
abline(v=2:K, col="gray")
#
# Se observa que k=10 da el máximo.
#                k= 3 es un máximo local. k=4 está próximo.
#
##########
#
# 4.  Multiples índices: 'NbClust'
#
# Previamente hay que descargar el paquete 'NbClust':
# > install.packages("NbClust")
#
library(NbClust)
#
distancias <- as.matrix(dist(datos, method="euclidean")^2)
NbClust(datos, diss=distancias, distance=NULL, method="kmeans")
#
#   Según los criterios empleados, 
#   hay un consenso: k=4 es el mejor número de clases,
#                  y k=2 es la siguiente opción.
#
###########################################################################
#
```
