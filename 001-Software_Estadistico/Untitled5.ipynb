```r
###########################################################################
#
# 'k-means', 'pam' ('Partitioning Around Medoids') y 'clara' (CLustering 
# LARge Applications) son algoritmos de clustering que definen particiones
# sobre un nmero predefinido 'k' de clases.
#
###########################################################################
#
# Estudio de la funcin 'pam()': 
#   http://en.wikipedia.org/wiki/K-medoids
#   http://en.wikibooks.org/wiki/Data_Mining_Algorithms_In_R/Clustering/Partitioning_Around_Medoids_%28PAM%29
#
# El algoritmo se basa en una matriz de desemejanzas o distancias.
#
# Los representantes de clase son siempre objetos del conjunto de datos.
#
# Cargar el paquete 'cluster' que contiene a 'pam()'
#
library(cluster) 
#
example(pam)
?pam
#
###########################################################################
#
# Introduccin de los datos: 'iris'
#
?iris
#
# Para una clasificacin no supervisada no se toma en cuenta
# la variable 'Species' de la planta (5a. columna)
#
datos0 <- iris[,1:4]
dim(datos0); nrow(datos0); ncol(datos0)
head(datos0)
summary(datos0)
#
# Preprocesamiento
#
# Estandarizar las variables?
#    Son homogneas (cm.); es razonable tratarlas tal cual.
#    Sin embargo, algunas medidas son ms grandes ('*.Length') que otras, 
#    y tratarlas tal cual equivaldra, en cierto sentido,
#    a analizarlas por su apariencia visual.
#
#    Estandarizar las variables equivale, en principio, a
#    poner todas las variables a un mismo nivel en el anlisis.
#    Esta es la opcin que se toma en el siguiente anlisis
#
# Para estandarizar las variables (media 0, varianza 1)
#
datosz <- scale(datos0)
#
# Verificacin:
#
for(j in 1:ncol(datosz)) print(c(mean(datosz[,j]), var(datosz[,j])))
#
# En adelante los datos sern los estandarizados
#
datos <- datosz
#
# si no fueran estandarizadas:
#
#                  datos <- datos0
#
###########################################################################
#
# Primer ensayo con 'pam()'
#
# Punto de partida: tabla de objetos x variables, 
#                   sobre la que se construye la matriz de distancias
#
# 'k' arbitrario
#
k <- 5
#
# 'pam()' no tiene ninguna componente aleatoria. Por tanto,
# no es necesario fijar ninguna semilla (a diferencia de 'kmeans()') 
# para obtener los mismos resultados en distintas ejecuciones del 
# siguiente cdigo:
#
pamx1 <- pam(datos, k, metric="euclidean", cluster.only=TRUE)
#
pamx1
#
# Caractersticas generales del objeto 'pamx1'
#
class(pamx1)
mode(pamx1)
attributes(pamx1)
#
###########################################################################
#
# Segundo ensayo con 'pam()'
#
# Punto de partida: una matriz de desemejanzas o distancias
#
# Calcular las distancias entre los objetos: 'dist()'
#
distancias <- dist(datos, method="euclidean", upper=TRUE, diag=TRUE)
#
k <- 5
#
pamx2 <- pam(distancias, k, cluster.only=TRUE)
#
pamx2
#
# Caractersticas generales del objeto 'pamx2'
#
class(pamx2)
mode(pamx2)
attributes(pamx2)
#
# Verificacin de que ambos puntos de partida,
# sea la matriz de datos o sea la matriz de distancias euclidianas,
# confluyen en el mismo resultado
#
table(pamx1, pamx2)
#
###########################################################################
#
# Tercer ensayo con 'pam()': forma usual 
#
# Punto de partida: una matriz de desemejanzas o distancias,
#                   tambin podra haber sido una matriz de datos   
#
distancias <- dist(datos, method="euclidean", upper=TRUE, diag=TRUE)
#
k <- 5
#
pamx <- pam(distancias, k) # cluster.only=FALSE es la opcin 'por defecto'
#
pamx
#
# Caractersticas generales del objeto 'pamx'
#
class(pamx)
mode(pamx)
attributes(pamx)
#
# Las caractersticas generales de estos resultados
# son distintas a las anteriores (cluster.only=TRUE).
#
# Hay ms informacin, que ser analizada en otro apartado
#
# pamx$medoids       Etiquetas de medoides en 'datos'
#
pamx$medoids
#
# pamx$id.med        Nmero de fila de medoides en 'datos'
#
pamx$id.med
#
pamx$objective
#
pamx$isolation
#
# Pertenencia de los objetos a las clases
#
pamx$clustering
#
# Composicin de las clases
#
compocl <- vector("list", k)
for(kcl in 1:k) compocl[[kcl]] <- which(pamx$clustering == kcl)
compocl
#
# 'pamxsize': tamaos de las clases  
#
pamxsize <- as.vector(table(pamx$clustering))
pamxsize 
#
# Verificacin de que la particin resultante es la misma
#
table(pamx1, pamx$clustering)
#
# Caractersticas de las clases
#
pamx$clusinfo 
#
# Valores del ndice de siluetas 
#
pamx$silinfo
#
plot(pamx)
#
#    Se observa que hay objetos cuyo valor 's(i) < 0'.
#               y podra procederse a intentar mejorarla,
#               tal y como se ha hecho con 'kmeans()'.
#
#    Se observa tambin que las clases conformadas no son 
#               las clases encontradas con 'kmeans()'.
#
###########################################################################
```
